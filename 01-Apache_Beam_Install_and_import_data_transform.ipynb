Incluindo transformações (Split)
Vendo o resultado percebe-se que foi adicionado o "u" na frente de cada valor do elemento. Esse "u" pode ser ignorado - (python unicode).

```
import apache_beam as beam

def SplitRow(element):
    return element.split(',')


p1 = beam.Pipeline()

attendance_count = (
    
   p1
    |beam.io.ReadFromText('dept_data.txt')
    |beam.Map(SplitRow)

    |beam.io.WriteToText('data/output_new')
)

p1.run()

# Sample the first 20 results, remember there are no ordering guarantees.
!{('head -n 20 data/output_new-00000-of-00001')}
```

# Olhando o mesmo exemplo, caso queira apresentar o resultado como uma Lista, pode-se usar FlatMap

```
import apache_beam as beam

def SplitRow(element):
    return element.split(',')


p1 = beam.Pipeline()

attendance_count = (
    
   p1
    |beam.io.ReadFromText('dept_data.txt')
    |beam.FlatMap(SplitRow)

    |beam.io.WriteToText('data/output_new')
)

p1.run()

# Sample the first 20 results, remember there are no ordering guarantees.
!{('head -n 20 data/output_new-00000-of-00001')}
```

# Continuando o exemplo anterior..Filter

```
import apache_beam as beam

def SplitRow(element):
    return element.split(',')

def filtering(record):
    return record[3]=='Accounts'

p1 = beam.Pipeline()

attendance_count = (
    
   p1
    |beam.io.ReadFromText('dept_data.txt')
    |beam.Map(SplitRow)
    |beam.Filter(filtering)
    |beam.io.WriteToText('data/output_new')
)

p1.run()

# Sample the first 20 results, remember there are no ordering guarantees.
!{('head -n 20 data/output_new-00000-of-00001')}
```

Continuando nosso exemplo ... contar a qtd de atendimentos, vamos usar lambda function

```
import apache_beam as beam

def SplitRow(element):
    return element.split(',')

def filtering(record):
    return record[3]=='Accounts'

p1 = beam.Pipeline()

attendance_count = (
    
   p1
    |beam.io.ReadFromText('dept_data.txt')
    |beam.Map(SplitRow)
    |beam.Filter(filtering)
    |beam.Map(lambda record: (record[1], 1))
    |beam.CombinePerKey(sum)

    |beam.io.WriteToText('data/output_new')
)

p1.run()

# Sample the first 20 results, remember there are no ordering guarantees.
!{('head -n 20 data/output_new-00000-of-00001')}
```
